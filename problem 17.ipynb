# Problem 17

If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?


NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

There will be four different types of numbers to compute based on number of digits. We will have to write out the numbers, separate them into lists of individual words, then count the letters in the words.

## One digit

We can define a function to find a single digit and count the letters. We will use this function to count the letters moving forward for multiple digits.

## Four digits (One thousand)

There is only one number we're concerned with here but in the interest of science I will write a function to include any four-digit number.

```python
# Define a function to return the string for the ones' digit

def ones(digit):
    if digit == '0':
        pass
    if digit == '1':
        return 'one'
    if digit == '2':
        return 'two'
    if digit == '3':
        return 'three'
    if digit == '4':
        return 'four'
    if digit == '5':
        return 'five'
    if digit == '6':
        return 'six'
    if digit == '7':
        return 'seven'
    if digit == '8':
        return 'eight'
    if digit == '9':
        return 'nine'

# Define a function to return the string for the tens' digit

def tens(digit):
    if digit == '2':
        return 'twenty'
    if digit == '3':
        return 'thirty'
    if digit == '4':
        return 'forty'
    if digit == '5':
        return 'fifty'
    if digit == '6':
        return 'sixty'
    if digit == '7':
        return 'seventy'
    if digit == '8':
        return 'eighty'
    if digit == '9':
        return 'ninety'
    if digit == '0':
        pass

# Define a function to return the string for the hundreds' digit

def hundreds(digit):
    if digit == '1':
        return 'one hundred'
    if digit == '2':
        return 'two hundred'
    if digit == '3':
        return 'three hundred'
    if digit == '4':
        return 'four hundred'
    if digit == '5':
        return 'five hundred'
    if digit == '6':
        return 'six hundred'
    if digit == '7':
        return 'seven hundred'
    if digit == '8':
        return 'eight hundred'
    if digit == '9':
        return 'nine hundred'

def teens(digit):
    if digit == 10:
        return 'ten'
    elif digit == 11:
        return 'eleven'
    elif digit == 12:
        return 'twelve'
    elif digit == 13:
        return 'thirteen'
    elif digit == 14:
        return 'fourteen'
    elif digit == 15:
        return 'fifteen'
    elif digit == 16:
        return 'sixteen'
    elif digit == 17:
        return 'seventeen'
    elif digit == 18:
        return 'eighteen'
    elif digit == 19:
        return 'nineteen'
    
```

```python
# Define a function to write out a number as a word

numlist = []

def write_num(num):
    """Write an integer as words"""
    numstring = str(num)
    numlength = len(numstring)
    if numlength == 1:
        numlist.append(ones(numstring))
    if numlength == 2:
        if num == 10:
            numlist.append('ten')
        elif num == 11:
            numlist.append('eleven')
        elif num == 12:
            numlist.append('twelve')
        elif num == 13:
            numlist.append('thirteen')
        elif num == 14:
            numlist.append('fourteen')
        elif num == 15:
            numlist.append('fifteen')
        elif num == 16:
            numlist.append('sixteen')
        elif num == 17:
            numlist.append('seventeen')
        elif num == 18:
            numlist.append('eighteen')
        elif num == 19:
            numlist.append('nineteen')
        elif numstring[-1] == '0':
            numlist.append(tens(numstring[-2]))
        else:
            numlist.append(tens(numstring[-2]))
            numlist.append(ones(numstring[-1]))
    if numlength == 3:
        if (numstring[-2] == '0') and (numstring[-1] == '0'):
            numlist.append(ones(numstring[-3]))
            numlist.append('hundred')
        if (numstring[-2] == '0') and (numstring[-1]) != '0':
            numlist.append(ones(numstring[-3]))
            numlist.append('hundred')
            numlist.append('and')
            numlist.append(ones(numstring[-1]))
        if (numstring[-2] == '1'):
            numlist.append(ones(numstring[-3]))
            numlist.append('hundred')
            numlist.append('and')
            numlist.append(teens(numstring[-2]))

for n in range(1, 201):
    write_num(n)

print(numlist)
```
