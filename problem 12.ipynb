The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

     1: 1
     3: 1,3
     6: 1,2,3,6
    10: 1,2,5,10
    15: 1,3,5,15
    21: 1,3,7,21
    28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?


```python
import math


# Define two functions to return a triangle number and its divisors.

def triangulate(n):
    """Return the nth triangle number"""
    addition = 0
    for k in range(1, n + 1):
        addition += k
    return addition

def divisors(num):
    """Return a list of divisors for any positive integer num"""
    divisors = [1, num]
    for k in range(2, int(math.sqrt(num)) + 1):
        if num % k == 0:
            divisors.append(num // k)
            divisors.append(k)
    divisors.sort()
    return divisors
```

```python
# Try getting triangle numbers by adding the next n value to the last item in a list and appending it

triangles = [0]
numdivisors = 0
n = 1
while numdivisors < 501:
    triangles.append(triangles[-1] + n)
    numdivisors = len(divisors(triangles[-1]))
    n += 1

print(triangles[-1])
print(numdivisors)
```
