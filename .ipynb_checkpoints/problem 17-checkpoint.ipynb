# Problem 17 - Number letter counts

If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

```python
# Let total = 11 for "one thousand"
# Create an alphabet string to count only letters

total = 11
alphabet = 'abcdefghijklmnopqrstuvwxyz'

for i in range(1, 1000):
    # Add letter counts for special-case numbers
    if i == 10:
        total += 3
    elif i == 11:
        total += 6
    elif i == 12:
        total += 6
    elif i == 13:
        total += 8
    elif i == 14:
        total += 8
    elif i == 15:
        total += 7
    elif i == 16:
        total += 7
    elif i == 17:
        total += 9
    elif i == 18:
        total += 8
    elif i == 19:
        total += 8
    else:
        # Convert numbers to strings and select digits by negative indices.
        num = str(i)
        digits = len(num)
        current = ''
        
        if num[-1] == '1':
            total += 3
            print(num)
        elif num[-1] == '2':
            total += 3
            print(num)
        elif num[-1] == '3':
            total += 5
            print(num)
        elif num[-1] == '4':
            total += 4
            print(num)
        elif num[-1] == '5':
            total += 4
            print(num)
        elif num[-1] == '6':
            total += 3
            print(num)
        elif num[-1] == '7':
            total += 5
            print(num)
        elif num[-1] == '8':
            total += 5
            print(num)
        elif num[-1] == '9':
            total += 4
            print(num)
        
        if digits == 2 or digits == 3:
            if num[-2] == '2':
                total += 6
                print(num)
            elif num[-2] == '3':
                total += 6
                print(num)
            elif num[-2] == '4':
                total += 5
                print(num)
            elif num[-2] == '5':
                total += 5
                print(num)
            elif num[-2] == '6':
                total += 5
                print(num)
            elif num[-2] == '7':
                total += 7
                print(num)
            elif num[-2] == '8':
                total += 6
                print(num)
            elif num[-2] == '9':
                total += 6
                print(num)
        
        if digits == 3:
            if num[-3] == '1':
                total += 13
                print(num)
            if num[-3] == '2':
                total += 13
                print(num)
            if num[-3] == '3':
                total += 15
                print(num)
            if num[-3] == '4':
                total += 14
                print(num)
            if num[-3] == '5':
                total += 14
                print(num)
            if num[-3] == '6':
                total += 13
                print(num)
            if num[-3] == '7':
                total += 15
                print(num)
            if num[-3] == '8':
                total += 15
                print(num)
            if num[-3] == '9':
                total += 14
                print(num)
            
            
# Subtract 27 for the missing "and" on "one hundred", "two hundred", etc.
print(total - 27)
        
```
